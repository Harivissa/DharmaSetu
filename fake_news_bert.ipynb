# AI Fake News Analysis using Language Models

This project explores the capabilities of a transformer-based language model in analyzing news headlines.  
The goal is to understand how pretrained language models interpret contextual information and emotional tone in news statements.

Language Model Used: DistilBERT (HuggingFace Transformers)

```python
from transformers import pipeline
import matplotlib.pyplot as plt
```

```python
classifier = pipeline("sentiment-analysis")

print("Language Model Loaded Successfully")
```

### Research Questions

1. How effectively can a pretrained language model analyze sentiment in news headlines?
2. Can sentiment analysis help identify sensational or emotionally manipulative news content?
3. What are the limitations of using general language models for fake news detection?

```python
## Language Model Selection

For this project, the DistilBERT language model from HuggingFace Transformers was used.  
DistilBERT is a lightweight version of BERT that retains strong natural language understanding capabilities while being computationally efficient.

The model is used to analyze news headlines and evaluate their contextual tone using sentiment classification.
```

```python
texts = [
    "Government launches new education reform improving schools.",
    "Breaking shocking miracle cure discovered that doctors hide.",
    "Scientists launch satellite successfully into orbit.",
    "Secret government conspiracy exposed urgent truth revealed."
]

results = []

for text in texts:
    result = classifier(text)[0]
    results.append(result)
    print("Headline:", text)
    print("Prediction:", result)
    print()
```

```python
## Research Questions

1. How effectively can a pretrained language model interpret news headlines?
2. Can sentiment analysis help identify emotionally manipulative or sensational news content?
3. What limitations arise when using general-purpose language models for misinformation detection?
```

```python
scores = [r['score'] for r in results]

plt.figure(figsize=(6,4))
plt.bar(range(len(scores)), scores)
plt.title("Language Model Confidence Scores")
plt.xlabel("News Samples")
plt.ylabel("Confidence Score")
plt.show()
```

## Exploration and Analysis

Multiple news headlines were tested to observe how the language model interprets different styles of writing.  
The model analyzes contextual patterns and provides sentiment predictions along with confidence scores.

The results show that the model focuses primarily on emotional tone and linguistic context.  
However, sensational headlines may still be classified as positive because the model is designed for sentiment analysis rather than factual verification.
